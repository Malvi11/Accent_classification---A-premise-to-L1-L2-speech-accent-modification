# Feature Extraction 

```python
import librosa 
import soundfile as sf
import glob
import IPython.display as ipd
from librosa import feature
import numpy as np
from sklearn.utils import shuffle
import librosa.display
import sklearn
```

```python
# audio = 'Recordingss/augmented_audio'
audio = r"C:\Recordingss\augmented_audio"
audio_files = glob.glob(audio + '/*.wav')
#print(audio_files)
```

```python
features = [librosa.feature.chroma_stft,
          librosa.feature.spectral_centroid,
          librosa.feature.spectral_bandwidth,
          librosa.feature.spectral_rolloff,
          librosa.feature.mfcc,
          librosa.feature.zero_crossing_rate
          ]

def get_feature_vector(y,sr): 
    
    feature_vec = [np.mean(funct(y,sr)) for funct in features]
    #feat_vect_ii = [ np.mean(funct(y)) for funct in fn_list_ii] 
    feature_vector = feature_vec
    return feature_vector
```

```python
fit = [] 

for file in audio_files:
    y,sr = librosa.load(file,sr=None)
    fv = get_feature_vector(y,sr)
    fit.append(fv)
    #print(fv)
```

```python
df=pd.DataFrame(fit)
#df.columns=c
df['English Type']=dfs['English Type'].astype(int)
df.rename(columns={0:'chroma_stft',1:'spectral_centroid',2:'spectral_bandwidth',3:'rolloff',4:'mfcc',5:'zero_crossing_rate'}, inplace=True)

for i in range(5):
    df = shuffle(df)
df.to_csv('Original_Features4.csv',index=False)
df.head()
```
