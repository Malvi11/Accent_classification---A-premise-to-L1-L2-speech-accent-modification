```python
import os
import glob
import shutil
 
source = r"/Users/malvi/OneDrive/Desktop/new/archive (1)/recordings"
destination =  r"/Users/malvi/OneDrive/Desktop/new/archive (1)/augmented_audio"
#
# gather all files
allfiles = glob.glob(os.path.join(source, '*.wav*'), recursive=True)
print("Files to move", allfiles)
 
# iterate on all files to move them to destination folder
for file_path in allfiles:
    dst_path = os.path.join(destination, os.path.basename(file_path))
    shutil.move(file_path, dst_path)
    print(f"Moved {file_path} -> {dst_path}")
```
