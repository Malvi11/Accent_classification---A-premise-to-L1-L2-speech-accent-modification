```python
import scipy.io.wavfile as wf
from scipy.fftpack import dct

dfs=pd.read_csv(r"C:\Recordingss\speakers_all.csv")
dfs.head()
```

```python
# audio = 'Recordingss/augmented_audio'
audio = r"C:\Recordingss\augmented_audio"
audio_files = glob.glob(audio + '/*.wav')
#print(audio_files)
```

```python
features = [librosa.feature.chroma_stft,
            librosa.feature.spectral_centroid,
            librosa.feature.spectral_bandwidth,
            librosa.feature.spectral_rolloff,
            librosa.feature.zero_crossing_rate,
    
          ]

def get_feature_vector(y,sr):
      
    #feature_vec = [np.mean(funct(y,sr)) for funct in features]
    feature_vec1 = [np.mean(funct(y, sr)) for funct in features]
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    for i in range(13):
        feature_vec1.append(np.mean(mfcc[i]))
    feature_vector = feature_vec1
    return feature_vector


    
```

```python
fit = [] 

for file in audio_files:
    y,sr = librosa.load(file,sr=None)
    fv = get_feature_vector(y,sr)
    fit.append(fv)
    print(fv)
```

```python
dfs=pd.read_csv(r"C:\Recordingss\speakers_all.csv")
dfs.head()

df=pd.DataFrame(fit)
df['English Type']=dfs['English Type'].astype(int)
df.rename(columns={0:'chroma_stft',1:'spectral_centroid',2:'spectral_bandwidth',3:'rolloff',4:'zero_crossing_rate',5:'mfcc_1',6:'mfcc_2',7:'mfcc_3',8:'mfcc_4',9:'mfcc_5',10:'mfcc_6',11:'mfcc_7',12:'mfcc_8',13:'mfcc_9',14:'mfcc_10',15:'mfcc_11',16:'mfcc_12',17:'mfcc_13'}, inplace=True)

for i in range(5):
    df = shuffle(df)
df.to_csv('Original_Features6.csv',index=False)
df.head()
```
